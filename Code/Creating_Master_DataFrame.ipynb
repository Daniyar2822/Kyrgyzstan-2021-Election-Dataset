```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
from tqdm import tqdm
```

In this notebook I wrangle raw data scraped from the website of Election Commission of Kyrgyzstan to create a dataset containing election results for Parliamentary Election of 2021 in the Kyrgyz Republic
Previosly I scraped raw tables for results of 37 district election commissions saved in 37 csv files  

```python
base_directory = '/Users/Daniyar/Documents/'
directory_path = '/Users/Daniyar/Documents/Election_Data/Single District Results/district_result'

relative_path = os.path.relpath(directory_path, base_directory)
```

```python
#creating variable for all csv files
csv_files = [f for f in os.listdir(relative_path) if f.endswith('.csv')]
```

```python
# Initialize an empty master dataframe
master_df2 = pd.DataFrame()
```

```python
#Creating one master dataframe for all data from csv files
for csv_file in tqdm(csv_files, desc="Processing CSV files", unit="file"):
    file_path = os.path.join(relative_path, csv_file)
    
    df = pd.read_csv(file_path)
    df.drop(['Unnamed: 0'], axis=1, inplace=True)
    df2 = df.T
    df2.columns = df2.iloc[0]
    #df3 = df2.rename(columns={'Кол-во голосов по кандидатам (Процент)' : 'Type'}, inplace=True)
    #df3 = df2.reset_index(inplace=True)
    
    # Concatenate the processed dataframe to the master dataframe
    master_df2 = pd.concat([master_df2, df2], ignore_index=False)
```

```python
#Checking if it works 
master_df2.shape
```

```python
master_df2.head()
```

```python
#Cleaning columns names 
master_df2.rename(columns={'Кол-во голосов по кандидатам (Процент)' : 'Type'}, inplace=True)
```

```python
#Dropping dublicated column
master_df2.drop('Кол-во голосов по кандидатам (Процент)', inplace=True)
```

```python
master_df2.head()
```

```python
#I need to reset index and then set it again and include Type columns as well 
master_df2.reset_index(inplace=True)
#df2.set_index(['index', 'Type'])
```

```python
master_df2.set_index(['index', 'Type'], inplace=True)
```

```python
#Using melt function reshaping dataframe grouping large number of columns to two: Cand and Votes
master_df2 = pd.melt(master_df2, var_name='Cand', value_name='Votes', ignore_index=False)
master_df2.head()
```

```python
#again resetting index
master_df2.reset_index(inplace=True)
```

```python
#Cleaning Type column removing redundant characters
master_df2['index'] = master_df2['index'].str.replace(r'Показать протокол(\.\d+)?', '', regex=True)
master_df2.head()
```

```python
#changing values in Type column, it is nesessary since later we create from them new columns
row_names = {'АСУ':'ASU',
             'АСУ.1':'ASU_per',
             'Ручной подсчет':'Hand',
            'Ручной подсчет.1': 'Hand_per'}
master_df2['Type'] = master_df2['Type'].replace(row_names)
```

```python
#Since our dataset contains such characters like character escape and percent sign
#Removing these characters to remain only numeric values
master_df2['Votes'] = master_df2['Votes'].str.replace('\xa0', '')
master_df2['Votes'] = master_df2['Votes'].str.replace('%', '')
```

```python
master_df2['Votes'] = master_df2['Votes'].astype('float')
```

```python
master_df2.rename(columns={'index':'UIK'}, inplace=True)
```

```python
#Next step is to create pivot table to show types of counting as columns 
df4 = master_df2.pivot_table(index=['UIK', 'Cand'], columns='Type', values='Votes', aggfunc=np.sum)
df4.head()
```

```python
df4.reset_index(inplace=True)
```

```python
#Next step is to clean up Cand column 
df4['Cand'] = df4['Cand'].str.title()
df4['Cand'] = df4['Cand'].str.replace('Уулу', 'уулу')
df4['Cand'] = df4['Cand'].str.replace('Кызы', 'кызы')
```

```python
#Since certain names are misspelled we need to match them

name_mapping = {'Каухар Тобокел кызы': 'Тобокел кызы Каухар', 'Эльвира Усен кызы': 'Усен кызы Эльвира', 
               'Исмаилова Тынчыгул Абдылдаевна': 'Исмаилова Тынчыгүл Абдылдаевна', 
                'Джанузакова Салтанат Смаилова': 'Джанузакова Салтанат Смаиловна', 
               'Бекназаров Руслан Руслан': 'Бекназаров Руслан Азимбекович',
               'Казматов Канатбек Кубанычбкович': 'Казматов Канатбек Кубанычбекович',
               'Момбекова Бактыгул Замирбековна':'Момбекова Бактыгүл Замирбековна',
                'Бейшекеева Калыгул Талантбекович': 'Бейшекеев Калыгул Талантбекович',
                'Торобеков Максатбек Асилбекович': 'Төрөбеков Максатбек Асилбекович',
                'Толобеков Омурзак Джолдошович': 'Төлөбеков Өмүрзак Джолдошович',
                'Алыбаева Жаныл Ишенбековна': 'Алыбаева Жаңыл Ишенбековна'}              

df4['Cand'] = df4['Cand'].replace(name_mapping)
            
```

```python
#I have other csv file containing information on party affiliation of all candidates
party_list = pd.read_csv('parties_list.csv')
party_list.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
#Merging two dataframes df4 and party_list 
#Our ultimate goal is to calculate number of voters which didn't choose the party
df5 = pd.merge(df4, party_list, on='Cand', how='left')
```

```python
df5 = df5[~df5['UIK'].str.startswith('№')]
```

```python
#Names of parties shall be unified
parties = {'«Кыргызстан Жашылдар Партиясы»': 'Жашылдар Партиясы',
          '«Ордо»': 'Ордо',
           '«Эл Үмүтү»': 'Эл Үмүтү',
           '«Аруузат-Эл Куту»': 'Аруузат-Эл Куту',
           '«Мекенчил Эл»': 'Мекенчил Эл',
           '«Азаттык»': 'Азаттык',
           '«Ынтымак»': 'Ынтымак',
           '«Патриотическая Партия Единения Кыргызстана»': 'ППЕК',
           '«Ишеним»': 'Ишеним',
           '«Улуттар Биримдиги»': 'Улуттар биримдиги',
           '«Ата-Мекен»': 'Ата-Мекен',
           '«Ыйман Нуру»': 'Ыйман Нуру',
           '«Социал-Демократтар»': 'Социал-демократтар',
           '«Бүтүн Кыргызстан»': 'Бүтүн Кыргызстан',
           '«Ата-Журт Кыргызстан»': 'Ата-Журт Кыргызстан',
           '«Улуу-Журт»': 'Улуу-Журт',
           '«Альянс»': 'Альянс',
           '«Багыт»': 'Багыт',
           '«Күчтүү Регион»': 'Күчтүү регион',
           '«Жашасын Кыргызстан»': 'Жашасын Кыргызстан',
           '«Легалайз»': 'Легалайз'}

df5['Cand'] = df5['Cand'].replace(parties)
```

Next steps are include the following:
1 - Creating dataframe containing only results of candidates 
2 - Creating separate dataframe with results of parties 

```python
#Identifing only candidates
candidates = df5.loc[df5['party'].notna()]
can = candidates['Cand'].unique()
candidates['Cand'].value_counts()
```

```python
#Creating dataframe specifically for candidates 
only_candidate_result = df5.loc[df5["Cand"].isin(can)].reset_index(drop=True)
```

```python
#Identifying party
party = df5.loc[df5['party'].isna()]
par = party['Cand'].unique()
```

```python
#Creating dataframe with results only for parties
only_parties_result = df5.loc[df5["Cand"].isin(par)].reset_index(drop=True)
only_parties_result.drop(['party', 'num', 'cand_status'], axis=1, inplace=True)
```

```python
only_cand_sum = only_candidate_result.groupby(['UIK','party'])[['ASU', 'Hand']].sum().reset_index()
```

```python
only_parties_result = pd.merge(only_parties_result, only_cand_sum, 
                         left_on=["UIK", "Cand"], right_on=["UIK", "party"], how="left")
```

```python
only_parties_result.drop(['party'], axis=1, inplace=True)
only_parties_result.rename(columns={'Cand': 'Party', 'ASU_x': 'ASU', 'Hand_x': 'Hand', 'ASU_y': 'Only_cand_ASU', 
                     'Hand_y': 'Only_cand_Hand'},
           inplace=True)
```

```python
#Adding column calculating numbers of votes without particular candidate choice 
only_parties_result['No_Cand_ASU'] = only_parties_result['ASU'] - only_parties_result['Only_cand_ASU']
only_parties_result['No_Cand_Hand'] = only_parties_result['Hand'] - only_parties_result['Only_cand_Hand']
```

```python
only_parties_result.fillna(0, inplace=True)
```

```python
only_parties_result = only_parties_result.astype({'ASU': 'int', 'Hand': 'int', 
                                                  'Only_cand_ASU': 'int', 'Only_cand_Hand': 'int',
                                                 'No_Cand_ASU': 'int', 'No_Cand_Hand': 'int'})
```

```python
only_candidate_result = only_candidate_result.astype({'ASU': 'int', 'Hand': 'int', 'num': 'int'})
```
